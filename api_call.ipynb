```python
import http.client
import json
import pprint

conn = http.client.HTTPSConnection("api.goldstandard.org")
payload = ''
headers = {
  'Content-Type': 'application/json'
}

conn.request("GET", "/projects?size=10000&gsCertified=true", payload, headers)
res = conn.getresponse()
gs_certified_project_list =  json.loads( res.read().decode())
```
